SimulaÃ§Ãµes de dados do Cordeiro Atacarejo



```python
# âœ… Bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# ðŸ“¦ Dados simulados do Cordeiro Atacarejo
dados = {
    "Data": pd.date_range("2024-01-01", periods=6, freq="M"),
    "Total_Vendas_Mensal": [1600000, 1620000, 1580000, 1655000, 1700000, 1750000],
    "Clientes_Atendidos": [42000, 41500, 40000, 43500, 44500, 46000],
    "Custo_Estoque": [720000, 740000, 710000, 735000, 755000, 780000],
    "Perdas_Pereciveis": [17000, 18500, 19000, 17200, 19800, 20000],
    "Funcionarios_Ativos": [165, 166, 167, 168, 170, 172],
    "Avaliacoes_Positivas": [84, 86, 87, 89, 91, 93],
}

df = pd.DataFrame(dados)
df["Ticket_Medio"] = df["Total_Vendas_Mensal"] / df["Clientes_Atendidos"]
df["%_Perdas"] = (df["Perdas_Pereciveis"] / df["Total_Vendas_Mensal"]) * 100
df["Vendas_por_Funcionario"] = df["Total_Vendas_Mensal"] / df["Funcionarios_Ativos"]
df.head()
```

```python
# ðŸ“ˆ Vendas Mensais
plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x="Data", y="Total_Vendas_Mensal", marker="o", color="green")
plt.title("Total de Vendas Mensal")
plt.ylabel("R$")
plt.xticks(rotation=45)
plt.show()
```

```python
# ðŸ“Š Vendas por FuncionÃ¡rio
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x="Data", y="Vendas_por_Funcionario", color="steelblue")
plt.title("Vendas por FuncionÃ¡rio")
plt.ylabel("R$")
plt.xticks(rotation=45)
plt.show()
```

```python
# ðŸŸ£ Pizza - AvaliaÃ§Ãµes Positivas
avaliacoes = df["Avaliacoes_Positivas"]
meses = df["Data"].dt.strftime('%b/%Y')
plt.figure(figsize=(8, 8))
plt.pie(avaliacoes, labels=meses, autopct='%1.1f%%', startangle=140)
plt.title("DistribuiÃ§Ã£o das AvaliaÃ§Ãµes Positivas")
plt.show()
```

```python
# ðŸ“‰ AnÃ¡lise: MÃªs com maior perda
maior_perda = df[df["Perdas_Pereciveis"] == df["Perdas_Pereciveis"].max()]
print("MÃªs com maior perda:")
print(maior_perda[["Data", "Perdas_Pereciveis"]])
```
